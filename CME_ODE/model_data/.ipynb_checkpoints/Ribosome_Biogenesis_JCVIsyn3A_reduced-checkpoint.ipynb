<h2> Use the Cells Below to Make the Simple Rate Laws for SSU and LSU Formation </h2>

```python
SSU_list = []
def getSSURF(speciesIDs,SSU_list):
    myStr = "$Par1"
    for ID in speciesIDs:
        if (("mRNA" in ID) or ("SU" in ID) or ("Ribosome" in ID) or ("L" in ID)):
            continue
        else:
            SSU_list.append(ID)
    for i in enumerate(SSU_list):
        print(i)
        addInt = i[0]+1
        myStr += "*$Sub" + str(addInt)
    return myStr
```

<h2> NOTE: Pass productBool as False if you would not like the reactants to be replenished (e.g. both reactants and products in the same reaction) with the formation of each ribosomal particle. </h2>

```python
'''
@MyDict Dictionary containing speciesIDs as keys, concentrations as values
@model the model object
@rxnIndx the index of the reaction
@productBool if True: all reactants are also products in the reaction
'''
def addSubs(MyDict,model,rxnIndx,productBool):
    i = 0
    for key,val in MyDict.items():
        i=i+1
        myStr="Sub" + str(i)
        model.addSubstrate(rxnIndx,myStr,key)
        if productBool == True:
            model.addProduct(rxnIndx,myStr,key)
    return 0
```

## Build the Solver object

```python
importlib.reload(desolver)

solv = desolver.solver.ModelSolver(model)

# We are not building for odeint (gives us more room to chose an integrator later).
# We are NOT using a jacobian, since we do not have the partial derivatives for all rate forms.
# We are building with Cython for speed, this is a big model.

solv.buildCall(odeint=False, useJac=False, cythonBuild=False, verbose=2)
```

```python
integrator = integrate.ode(solv, jac=None)#.set_integrator('vode',solver='adams',min_step=1e-6)
#print(type(solv))
#integrator.set_integrator(min_step=0.1,max_step=1)
integrator.set_initial_value(model.getInitVals())

totalTime = 7200 #3600
step = 0.1

print("Initial concentration: ", model.getInitVals())
print('\n')

results = np.empty((0,len(model.getInitVals())), float)

while integrator.successful() and integrator.t < totalTime:
    currConcentration = integrator.integrate(integrator.t + step)
    #if (integrator.t%10.0 == 0.0):
    print(integrator.t)
    #print(integrator.t, currConcentration)
    results = np.append(results, [np.asarray(currConcentration)], axis=0 )

print('\n')
print("Final concentration at time " + str(totalTime) + ":", currConcentration)
```

```python
# We can calculate the reaction fluxes explicitly, instead of the "deltas" for metabolite concentrations.
solv.calcFlux(0, np.asarray(model.getInitVals()) )
```
